# India Rainfall Analysis

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("rainfall in india 1901-2015.csv")
data.info()
```

```python
data.head(1500)
```

```python
data.describe()
```

```python
annunal_rainfall = data.groupby("YEAR").sum()['ANNUAL']
annunal_rainfall.plot(title="ANNUAL RAINFALL", figsize=(15,10))
```

Figure shows annual rainfall in India from year 1901 to 2015

```python
data[['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("YEAR").sum().plot(figsize=(20,12));
```

Monthly analysis of rainfall in India from year 1901 to 2015

```python
data[['YEAR','Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].groupby("YEAR").sum().plot(figsize=(13,8));
```

Span of Jun-Sep has maximum rainfall records.

```python
data[['SUBDIVISION', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("SUBDIVISION").mean().plot.barh(stacked=True,figsize=(13,8));
```

Graph shows top 3 subdivisions having high average rainfall:
1. ARUNACHAL PRADESH
2. COASTAL KARNATAKA
3. KOKAN & GOA

```python
data[['SUBDIVISION', 'Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].groupby("SUBDIVISION").sum().plot.barh(stacked=True,figsize=(16,8));
```

```python
data[['SUBDIVISION', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("SUBDIVISION").max().plot.barh(stacked=True, figsize=(16,8));
```

Graph shows Maximum rainfall for specific month registered in ARUNACHAL PRADESH.

# Predicting Average Rainfall in Bihar

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Ridge
```

```python
df = data.groupby('SUBDIVISION')['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','NOV','DEC']
b_data=df.get_group(('BIHAR'))
b_data.head()
```

```python
b_data=b_data.melt(['YEAR']).reset_index()
b_data.head()
```

```python
df= b_data[['YEAR','variable','value']].reset_index().sort_values(by=['YEAR','index'])
df.head()
```

```python
df.columns=['INDEX','YEAR','MONTH','AVG_RAINFALL']
df.head()
```

```python
d={'JAN':1,'FEB':2,'MAR' :3,'APR':4,'MAY':5,'JUN':6,'JUL':7,'AUG':8,'SEP':9,
   'OCT':10,'NOV':11,'DEC':12}
df['MONTH']=df['MONTH'].map(d)
df.head(12)
```

```python
df['DATE']=pd.to_datetime(df.assign(Day=1).loc[:,['YEAR','MONTH','Day']])
df.head(12)
```

```python
X = df['DATE'].values.reshape(-1,1)
y = df['AVG_RAINFALL'].values.reshape(-1,1)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```
